# Texto Escolhido

O texto escolhido foi a biografia de John Ronald Reuel Tolkien, autor, professor universitário, filólogo e inspiração para diversas obras da literatura e cinema moderno.

# Carregamento e Tratamento dos Dados

```python
import urllib.request  

texto = []

for line in urllib.request.urlopen('https://raw.githubusercontent.com/digonfernan/exercicio-lei-de-zipf/main/biografia-tolkien.txt'): 
    texto.append(line.decode('utf-8'))

for i in range(len(texto)):
    # Para ter singularidade
    texto[i] = texto[i].lower()
    # Elimina carcteres indesejáveis
    texto[i] = texto[i].replace('\n','')
    texto[i] = texto[i].replace('.','')
    texto[i] = texto[i].replace(',','')
    texto[i] = texto[i].replace('(','')
    texto[i] = texto[i].replace(')','')
    texto[i] = texto[i].replace('?','')
    texto[i] = texto[i].replace('\'','')
```

# Elaboração do Dicionário

```python
# Dicionário vazio
dic = {}

for line in texto:
  # Converte para caixa baixa
  line = line.lower()
  # Separa as palavras
  words = line.split()

  # Loop para contar as palavras
  for word in words:
    if word not in dic.keys(): 
      dic[word] = 1
    else:
      dic[word] =  dic[word] + 1

print(dic)    
```

# Resultados

```python
# Plot
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

df = pd.DataFrame(dic.items(),  columns=['word', 'count']).sort_values('count',ascending=False)
# Conta apenas palavras com ocorrência maior que 4 vezes
df = df[df['count'] > 4]

plt.figure(figsize=(24,10))
mpl.style.use(['seaborn'])
sns.barplot(x=df.word,y=df['count'])
plt.xticks(rotation=90) 

plt.show()
```
